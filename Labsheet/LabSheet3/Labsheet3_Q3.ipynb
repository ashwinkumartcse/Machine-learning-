```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
import pandas as pd
import numpy as np
data=pd.read_csv("C:/Users/Deepika Palaniappan/Documents/machine learning/LabSheet3/car.data",sep=',',names=(['buying','maintainence','doors','persons','luggage_boot','safety','class']))
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data1=data.replace({'buying':{'vhigh': 4, 'high': 3,'med':2,'low':1}})
data1.head()
```

```python
data2=data1.replace({'maintainence':{'vhigh': 4, 'high': 3,'med':2,'low':1}})
data2.head()
```

```python
data3=data2.replace({'luggage_boot':{'small': 1, 'med': 2,'big':3}})
data3.head()
```

```python
data4=data3.replace({'safety':{'low': 1, 'med': 2,'high':3}})
data4.head()
```

```python
data5=data4.replace({'class':{'unacc': 1, 'acc': 2,'good':3,'vgood':4}})
data5.head()
```

```python
data6=data5.replace({'persons':{'2': 2, '3': 3,'4':4,'5more':'5','more':'5'}})
data6.head()
```

```python
final=data6.replace({'doors':{'2': 2, '3': 3,'4':4,'5more':'5'}})
final.head()
```

```python
X = final.drop(['class'],axis=1)
y = final['class']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=2)
```

```python
cars = LogisticRegression()
cars.fit(X_train,y_train)
```

```python
y_pred=cars.predict(X_test)
```

```python
print("Accuracy of Logistic Regression model is:",
metrics.accuracy_score(y_test, y_pred))
```
